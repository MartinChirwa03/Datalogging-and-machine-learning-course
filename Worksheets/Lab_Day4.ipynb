```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python



file_path = r"C:\Users\pc\OneDrive\Desktop\archive\lodges.csv" 
df = pd.read_csv(file_path)
df.info
```

```python
# use some of the following to inspect your loaded data - check Pandas documentation(DataFrames) for more


location_col = 'location'  


df = df.dropna(subset=[location_col])


def clean_location(name):
    name = name.strip().lower()
    name = name.replace(".", "") 
    name = name.replace(" city", "")  
    return name.title()

df[location_col] = df[location_col].apply(clean_location)


location_counts = df[location_col].value_counts()


```

4. Based on your inspection, you now need to identify the relevant parts of the dataset that address your research question. 
Select the necessary data, and visualize it using an appropriate plot. 

```python
plt.figure(figsize=(12, 6))
location_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Location')
plt.ylabel('Number of Hotels')
plt.title('Number of Hotels per Location')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
